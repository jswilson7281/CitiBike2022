```python
import pandas as pd
import os 
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data_sample_temps.csv')
```

```python
lat_long=df[['start_station_name','end_station_name','start_lat','start_lng','end_lat','end_lng']].drop_duplicates(subset=['start_station_name','end_station_name'])
```

```python
lat_long
```

```python
df['value'] = 1
df_group = df.groupby(['start_station_name', 'end_station_name'])['value'].count().reset_index()
```

```python
df_group
```

```python
df_plot=lat_long.merge(df_group,on=['start_station_name', 'end_station_name'])
df_plot
```

```python
df_plot=df_plot.nlargest(n=100,columns='value')
```

```python
df_plot.to_csv("df_plot.csv")
```
