```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_sample_temps.csv')
```

```python
df
```

```python
df['member_casual'].value_counts().plot(kind = 'bar')
```

```python
df['value'] = 1
df_t = df.groupby('start_station_name')['value'].count().reset_index()
```

```python
top20 = df_t.nlargest(20, 'value')
top20
```

```python
sns.barplot()
```

```python
sns.barplot(data = top20, x = 'value', y = 'start_station_name' )
```

```python
sns.barplot(data = top20, x = 'value', y = 'start_station_name', palette=("Blues")) 
plt.title("Most frequent start stations")
plt.show()
```

```python
df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = 'date', how = 'outer')
```

```python
fig,ax = plt.subplots(figsize=(10, 5))
sns.lineplot(data = df['bike_rides_daily'], color = "b")
ax.set_xlabel("Year 2018", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily", color = "navy", fontsize = 14) # set y-axis label
ax2 = ax.twinx()
sns.lineplot(data = df['avgTemp'], color="r", ax=ax2)
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 2018', fontsize = 18)
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(x = 'rideable_type', y = 'avgTemp', data = df, palette = "Blues_r")
plt.show()
```

```python
grid = sns.FacetGrid(df, col = "rideable_type")
grid.map(sns.histplot, "avgTemp", bins = 15)
```
